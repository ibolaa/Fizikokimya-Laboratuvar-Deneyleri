```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

BOYLE MARIOTTE 

```python
data = {"p":[0.9960, 0.9446, 0.9191, 0.8879, 0.8743, 0.8382],"v":[0.05,0.052,0.054,0.056,0.058,0.060]}
df = pd.DataFrame(data)
df["1/v"] = df["v"].apply(lambda x: x**(-1))
df
```

```python
plt.figure(figsize=(12,4))

sns.regplot(x=df["1/v"], y=df["p"], ci=None, scatter_kws={"s": 50},line_kws={"linewidth": 0.5}, marker="+", color="r")

plt.grid(axis="both", which="both", color="black",linestyle="-", linewidth=0.1)

plt.xlim(16,20.5)
plt.ylim(0.82,1.02)
plt.xlabel("1/V (L^-1)")
plt.ylabel("P (atm)")

np.polyfit(df["1/v"],df["p"],deg=1)
m = np.polyfit(df["1/v"],df["p"],deg=1)[0]
b = np.polyfit(df["1/v"],df["p"],deg=1)[1]
print("y = {m}*x + {b}".format(m=m,b=b))
print("P*V = n*R*T = {m} ".format(m=m))
```

GAY LUSSAC ABSOLUTE ZERO PREDICTION

```python
data = {"T":[23,29,36,43,49],"ml":[45,46,47,48,49]}
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(12,4))


np.polyfit(df["T"],df["ml"],1)

m = np.polyfit(df["T"],df["ml"],1)[0]
b = np.polyfit(df["T"],df["ml"],1)[1]


l = ((np.linspace(0,49,5))-np.polyfit(df["T"],df["ml"],1)[1])/np.polyfit(df["T"],df["ml"],1)[0]

plt.plot(l, np.linspace(0, 49, 5),color="r",linewidth=0.8,)
plt.grid(axis="both", which="both", color="black",linestyle="-", linewidth=0.1)
plt.xlim(-274.4848426915337,50)
plt.ylim(0,50)

print("y = {m}*x + {b}".format(m=m,b=b))
print("predicted absolute zero is {abs} °C".format(abs=-b/m))
```
